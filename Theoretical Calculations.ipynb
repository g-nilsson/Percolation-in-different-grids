```python
import matplotlib.pyplot as plt
import numpy as np

# Cobweb plot

p_vals = np.linspace(0, 1, 1000)
f = lambda x: x**4 + 4*x**3*(1-x) + 2*x**2*(1-x)**2
plt.figure(dpi=300)

plt.plot(p_vals, f(p_vals), label="f(p)=Φ(p)")
plt.plot(p_vals, p_vals, label="f(p)=p")
plt.scatter([0, 0.61803398874989, 1], [0, 0.61803398874989, 1], label='critical points')
plt.legend()
plt.xlabel('p_n - probability of percolation at scale n')
plt.ylabel('p_2n - probability of percolation at scale 2n')

val = 0.5
for _ in range(4):
    plt.arrow(val, f(val), f(val)-val, 0, head_width=0.025, length_includes_head=True)
    plt.arrow(f(val), f(val), 0, f(f(val))-f(val), head_width=0.025, length_includes_head=True)
    val = f(val)

plt.savefig('square_cobweb.png')
```
